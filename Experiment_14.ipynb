Experiment 14: Data binning and Data Formatting using Python


```
import pandas as pd
import numpy as np

data ={
'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
'Price':[55000,20000,30000,2000,8000,45000],
'Units_sold':[25,60,35,80,50,15]
}
df = pd.DataFrame(data)
print(df)
```

```
# Data Binning
bins1 = [0,10000,30000,60000]
labels1 = ['Low','Medium','High']
df['Price_Category'] = pd.cut(df['Price'],bins=bins1,labels=labels1)

print(df)
```

```
# Binning Units sold
bins2 =[0,30,60,100]
labels2 =['Low Sales','Medium Sales','High Sales']

df['Sales_Category'] = pd.cut(df['Units_sold'],bins=bins2,labels=labels2)
print(df)
```

```
# Data Formatting
# Convert Data Type

print(df.dtypes)

df['Units_sold']= df['Units_sold'].astype(float)
print(df.dtypes)
df
```

```
# Convert text to uppercase for the 'Product' column
df['Product'].str.upper()

print(df)
```

```
# Rounding Values
df['Price'] = df['Price'].round(2)
print(df)
```

```
# Additional
# Sort Data & overwriting it.
df.sort_values(by='Price', ascending=False, inplace=True)
df
```

```
# Display Unique Values
print(df['Price_Category'].unique())
```

```
# Dataset: Food Delivery orders
import pandas as pd
import numpy as np

data ={
'Order_ID': ['01','02','03','04','05','06','07','08'],
'Order_Value':[250,450,120,600,300,150,500,220], # Added np.nan to make it 8 elements
'Delivery_Time':[30,45,25,60,35,20,50,28], # Formatted as a list of 8 strings
'Distance_km':[5,8,3,10,6,2,9,4]
}
df = pd.DataFrame(data)
print(df)
```

```
# Binning Order Value
bins =[0,200,400,700]
labels ={'Low Order','Medium Order','High Order'}
df['Order_Category'] = pd.cut(df['Order_Value'],bins=bins,labels=labels)

print(df)
```

```
# Binning delivery time
bins2 = [0,30,45,70]
labels2 = ['Fast','Normal','Slow']

df['Delivery_Speed'] = pd.cut(df['Delivery_Time'], bins=bins2, labels=labels2)

print(df)
```

```
# Binning distance_km
bins3 = [0,5,8,10]
lables3 = ['Short','Medium','Long']
df['Delivery_Distance'] = pd.cut(df['Distance_km'], bins=bins3, labels=lables3)

print(df)
```

```
# Count of categories
print(df['Order_Category'].value_counts())
print(df)
```

```
# Sorting
print(df.sort_values(by=['Order_Value']))
```

```
# Converting dataset into uppercase
df['Order_Value'] = df['Order_Value'].astype(str).str.upper()
print(df)
```
